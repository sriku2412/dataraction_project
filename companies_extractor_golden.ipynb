```python
import json

def extract_keys_from_dict(d, parent_key='', keys_list=[]):
    for k, v in d.items():
        new_key = f"{parent_key}.{k}" if parent_key else k
        if isinstance(v, dict):
            extract_keys_from_dict(v, new_key, keys_list)
        elif isinstance(v, list):
            if v and isinstance(v[0], dict):
                extract_keys_from_dict(v[0], new_key, keys_list)
        else:
            keys_list.append(new_key)
    return keys_list



response_dict = json.loads(response.text)
keys_list = extract_keys_from_dict(response_dict)
print(keys_list)
```
